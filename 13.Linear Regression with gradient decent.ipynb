Linear Regression using Gradient Descent

```python
from PIL import Image
i=Image.open('gradient decentt.jpg')
i
```

```python
m=0  #gradient
c = 0 # y intercept
l = 0.0001 #learning rate
n = int(len(x)) #number of datapoints
iterations = 1000
```

```python
for i in range(iterations):
    pred = (m*x) + c
    der_m = (-2/n) *sum(x*(y-pred))
    der_c = (-2/n) *sum((y-pred))
    m =m-(l*der_m)
    c =c-(l*der_c)
print(m , c)
```

```python
for i in range(iterations):
    pred = (m*x) + c
    der_m = (-2/n) *sum(x*(y-pred))
    der_c = (-2/n) *sum((y-pred))
    m =m-(l*der_m)
    c =c-(l*der_c)
    print(m,c)
```

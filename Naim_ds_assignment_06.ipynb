# One hot Encoding

```python
df6.head()
```

```python
dummi = pd.get_dummies(df6['region'])
dummi.head()
```

```python
dumi_drop = pd.get_dummies(df6['region'],drop_first=True)
dumi_drop.head()
```

```python
df6 =df6.drop('region',axis=1) #axis=1 means column,drop ,,axis=0 means row delete
df6.head()
```

# Loop for One hot Encoding

```python
df7.head()
```

```python
for col in df7.columns:
    if is_numeric_dtype(df7[col]):
        continue
    else:
        one=pd.get_dummies(df7[col], drop_first=True, prefix='one hot col')
        df7.drop(df7[[col]], axis=1, inplace=True)
        df7=pd.concat([df7, one], axis=1)
```

```python
df7.head()
```
